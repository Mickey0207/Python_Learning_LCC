```python
import matplotlib.pyplot as plt

x=[]
y=[1,2,0.8,1.5,0.7,3]
for i in range(1,7):
    x.append(i)
print(x)   

plt.title('income')
plt.xlabel('Month')
plt.ylabel('10k nts')
plt.plot(x,y,'r^--')
```

```python
import matplotlib.pyplot as plt
money=100000
rate=0.2
n=7
x=[]   #串列初始條件
y=[]  

for i in range(1,n):
    money=money*(1+rate)
    x.append(i)
    y.append(round(money))
print(x)    
print(y)

plt.bar(x,y,width=1)
plt.show()
```
